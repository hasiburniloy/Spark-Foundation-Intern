# Unsupervised ML 
### Here we will findout the optimum cluster of the iris dataset 

## Importing llibrary

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.cluster import KMeans
from sklearn import datasets
plt.style.use('seaborn-colorblind')
```

## Loading dataset

```python
iris = datasets.load_iris()
data=pd.DataFrame(iris.data,columns=iris.feature_names)
data.head()
```

## Visualizing the data

```python
pd.plotting.scatter_matrix(data,figsize=(8,8),diagonal='kde');
```

##  Finding the optimum Cluster

```python

x=data.iloc[:,0:].values
w=[]
for i in range(1,10):
    
    km=KMeans(n_clusters=i,init='k-means++',
    n_init=10,
    max_iter=500,
              random_state=0) #setting the hyperparameter of the model
    
    km.fit(x) #fitting the value in the model
    w.append(km.inertia_) #appending the loss value in a list
plt.plot(range(1,10),w,"-g")
plt.title("Elbow method")
plt.xlabel("Number of K")
plt.ylabel("Loss")
plt.plot(3,95,'r|')
plt.show()
```

The elbow method is use to find the best <b>K</b> value for the data. Here 3 cluster will be better for the dataset with 500 iteration.

Applying the cluster on the data

```python
km=KMeans(n_clusters=3,init='k-means++',
    n_init=10,
    max_iter=500,
              random_state=0)
y=km.fit_predict(x)
```

```python
%matplotlib notebook
plt.figure(figsize=(8,8))
color=['Red','blue','green']
marker=["p","P","*"]
for i in range(0,3):
    plt.scatter(x[y==i,0],x[y==i,1],s=100,
    c=color[i],marker=marker[i],label="Cluster:{}".format(i+1))

plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:,1],s = 100, c = 'Black', label = 'Centroids')
plt.title("Visualization of the cluster")
plt.legend()
```

So here we can see the clustering of the data.
